```python
import pandas as pd
import numpy as np
df = pd.read_csv('Database Anggota Ikama Bandung (Responses) - Form Responses 1.csv')
```

```python
df
```

```python
df.columns
df.rename(columns={'Timestamp':'Timestamp','Nama Lengkap ':'NAMA','Tempat, Tanggal Lahir (Contoh: Medan, 08 November 1997)':'TTL','Daerah Asal Kota/Kabupaten (Contoh: Kota Medan)':'ASAL',
                  'Angkatan Matauli (Contoh: 23)':'ANGKATAN','Pekerjaan Saat Ini (Contoh: Mahasiswa)':'PEKERJAAN','Asal Universitas/Instansi -Jurusan (Contoh : ITB-Teknik Informatika)':'UNIVERSITAS',
                  'Riwayat Pendidikan Lanjut (Bagi yang melanjutkan studi S2 atau S3) ':'RIWAYAT PENDIDIKAN LANJUT','Alamat Lengkap di Bandung (Contoh : Jl. Cisitu Lama XI No.3, Kecamatan Coblong, Kota Bandung. Jika diluar Bandung isi "-")':'ALAMAT BDG',
                  'Email aktif':'EMAIL','Nomor WhatsApp':'NOMOR WA','Id Line (Jika tidak ada isi "-")':'ID LINE','Kontak Darurat - Hubungan (Contoh: 082361759188-Ibu)':'KONTAK DARURAT',
                  'Ide, Saran, dan Masukan untuk Kepengurusan Ikama Bandung Periode 2022-2024':'IDE','Foto terbaru kamu sendirian untuk keperluan divisi kreatif':'IDE',
                  'Foto terbaru kamu sendirian untuk keperluan divisi kreatif':'LINK FOTO','Perkiraan Wisuda/Kelulusan - Bagi Angkatan 22, 21, & 20 (Cth : April 2022)':'WISUDA','Email Address':'Email Address'},inplace=True)
df.drop('Timestamp',axis=1,inplace=True)
df.drop('Email Address',axis=1,inplace=True)
```

```python
df.isna()
df.isnull().sum()
df['RIWAYAT PENDIDIKAN LANJUT'].fillna('-',inplace=True)
```

```python
ttl_sep=df['TTL'].str.split(',',expand=True)
ttl_sep.iloc[4,1]='22 September 1993'
ttl_sep.iloc[4,0]='-'
```

```python
tgl = ttl_sep[1]
bulan_lahir = ttl_sep[1].str.split(' ',expand=True)
bulan_lahir.iloc[4,1] = '22'
bulan_lahir.iloc[4,2] = 'September'
bulan_lahir.iloc[27,1] = '11'
bulan_lahir.iloc[27,2] = 'Juni'
bulan_lahir.drop([0,3],axis=1,inplace=True)
bulan_lahir[2]=bulan_lahir[2].str.capitalize()
bulan_lahir
```

```python
df['TANGGAL LAHIR'] = bulan_lahir[1]
df['BULAN LAHIR'] = bulan_lahir[2]
```

```python
df
```

```python
new_df = df.sort_values(['BULAN LAHIR','TANGGAL LAHIR'])
new_df.to_excel('DATABASE IKAMA BANDUNG per APRIL 2022.xlsx')
```
